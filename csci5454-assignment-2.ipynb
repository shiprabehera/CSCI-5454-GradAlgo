## Problem 1: Fixed Point in an Array (15 points)

__(A, 5 points)__ A fixed point of an array $A$, if it exists, is an index $i$ such that $A[i] = i$.
Given a _sorted_ array $A$ of _distinct_ integers, return the index of the fixed point if one exists, or otherwise, return `-1` to signal that no fixed point exists. Your algorithm must be as efficient as possible.

```python
def find_fixed_point(a):
    # YOUR SOLUTION HERE
    left = 0
    right = len(a) - 1
    while (left <= right):
        mid = int(left + (right-1)/2)
        if a[mid] == mid:
            return mid
        elif a[mid] < mid:
            left = mid + 1
        else:
            right = mid - 1
    return -1       
    # END
```

## Problem 2: FFT problem (15 points)

Consider a sequence of numbers $(a_0, \ldots, a_{n-1})$ and let $(A_0, A_1, \ldots, A_{n-1})$ be its
discrete fourier transform: 

$$ A_j = \sum_{k=0}^{n-1} a_k (\omega_n^{jk})$$

where $\omega_n = e^{\left( \frac{2 \pi i }{n} \right)}$


__(A, 7 points)__  Consider the Fourier transform of the circular permutation $(a_1, a_2, \ldots, a_{n-1}, a_0)$. Let us denote it as $( \hat{A}_0, \ldots, \hat{A}_{n-1})$. Find an expression for $\hat{A}_j$ in terms of $A_j$. **Hint** Use the fact that $\omega_n^n = 1$.

__(B, 8 points)__ Generalize part (A) for the circular permutation $(a_{p}, \ldots, a_{n-1}, a_0, \ldots, a_{p-1})$. Use the fact that $\omega_n^{n+j} = \omega_n^j$.

### Answer (15 lines)

__(A)__
$ A_j = \sum_{k=0}^{n-1} a_k (\omega_n^{jk})$ <br/>
Using the above Discrete Fourier Transform, for circular permutation $(a_1, a_2, \ldots, a_{n-1}, a_0)$, <br/>
we get,  $\hat{A}_j = a_1\omega_n^{j.0} + a_2\omega_n^{j.1} + \cdots + a_{n-1}\omega_n^{j(n-2)} + a_0\omega_n^{j(n-1)}$<br/>
$$\implies \hat{A}_j = a_0\omega_n^{j(n-1)} + a_1\omega_n^{j.0} + a_2\omega_n^{j.1} + \cdots + a_{n-1}\omega_n^{j(n-2)} $$ <br/><br/>
DFT of sequence $(a_0, \ldots, a_{n-1})$, <br/>
$A_j = a_0\omega_n^{j.0} + a_1\omega_n^{j.1} + a_2\omega_n^{j.2} + \cdots + a_{n-1}\omega_n^{j(n-1)} $ <br/><br/>
Since $\omega_n^n = 1$ and $\omega_n^0 = 1$ and $(\omega_n^{n})^{j} = 1$, we can write the first term, $a_0\omega_n^{j.0}$ as $a_0\omega_n^{jn}$ <br/><br/>
$\implies A_j = a_0\omega_n^{jn} + a_1\omega_n^{j.1} + a_2\omega_n^{j.2} + \cdots + a_{n-1}\omega_n^{j(n-1)} $ <br/><br/>
Taking out $\omega_n^{j}$ from each term, we get<br/>
$A_j = a_0\omega_n^{j(n-1)}\omega_n^j + a_1\omega_n^{j.0}\omega_n^j + a_2\omega_n^{j.1}\omega_n^j + \cdots + a_{n-1}\omega_n^{j(n-2)}\omega_n^j $ <br/><br/>
$\implies A_j = \omega_n^j(a_0\omega_n^{j(n-1)} + a_1\omega_n^{j.0} + a_2\omega_n^{j.1} + \cdots + a_{n-1}\omega_n^{j(n-2)}) $ <br/><br/>
$$\implies \frac{A_j}{\omega_n^j} = a_0\omega_n^{j(n-1)} + a_1\omega_n^{j.0} + a_2\omega_n^{j.1} + \cdots + a_{n-1}\omega_n^{j(n-2)} $$ <br/>
The right side of this equation is equal to $\hat{A}_j$ which we derived before.<br/>
Hence, $$\hat{A}_j = \frac{A_j}{\omega_n^j}$$

__(B)__
Using the above Discrete Fourier Transform, for circular permutation $(a_{p}, \ldots, a_{n-1}, a_0, \ldots, a_{p-1})$, <br/>
we get,  $\hat{A}_j = a_p\omega_n^{j.0} + a_{p+1}\omega_n^{j.1} + \cdots + a_{n-2}\omega_n^{j(n-2-p)} + a_{n-1}\omega_n^{j(n-1-p)}+ \cdots + a_0\omega_n^{j(n-p)} + a_1\omega_n^{j(n-p+1)} + \cdots + a_{p-1}\omega_n^{j(n-1)}$<br/><br/>
$\implies \hat{A}_j = \sum_{k=p}^{n-1} a_k (\omega_n^{j(k-p)}) + \sum_{k=0}^{p-1} a_k (\omega_n^{j(n-p+k)})$ <br/><br/>
$\hat{A}_j = \sum_{k=p}^{n-1} a_k (\omega_n^{jk})(\omega_n^{-jp}) + \sum_{k=0}^{p-1} a_k (\omega_n^{j(n+k)})(\omega_n^{-jp})$ <br/><br/>
$\omega_n^{-jp}$ is a constant inside the summation, so we can take that out from both terms<br/>
$\hat{A}_j = (\omega_n^{-jp}) (\sum_{k=p}^{n-1} a_k (\omega_n^{jk}) + \sum_{k=0}^{p-1} a_k (\omega_n^{j(n+k)}))$ <br/><br/>

Since, $\omega_n^n = 1, \omega_n^{n+j} = \omega_n^n\omega_n^j = \omega_n^j$, we can write $(\omega_n^{n+k})^{j}$ as $\omega_n^{jk}$<br/>
$\implies \hat{A}_j = (\omega_n^{-jp}) (\sum_{k=p}^{n-1} a_k (\omega_n^{jk}) + \sum_{k=0}^{p-1} a_k (\omega_n^{jk}))$ <br/><br/>
$\implies \hat{A}_j = (\omega_n^{-jp}) (\sum_{k=0}^{n-1} a_k (\omega_n^{jk}) $ <br/><br/>
Hence, $$\hat{A}_j = \frac{A_j}{\omega_n^{jp}}$$

## Problem 4 (15 points)

Show that the summation 

$$ \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}} = n a_{\color{red}{m}} $$
for $m = 0, \ldots, n-1$. Here $\omega$ is the $n^{th}$ root of unity and $\overline{\omega} = \omega^{-1}$.
This establishes the fact that the inverse of the FFT is just the same algorithm using $\overline{\omega}$ instead of $\omega$, as discussed in class.

**Hints** Do not panic!!  

1. Bring the $(\overline{\omega})^{j \color{red}{m}}$ term into the summation, using the fact that $\overline{\omega} = \omega^{-1}$.

2. Next, change the order of summation taking each $a_k$ out in common.

3. Use the geometric series summation and you will see all coefficients cancel to $0$ other than that for $a_m$.

4. Arrive at the answer.

## Solution (Expected Length: 15 lines)


$ \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  (\overline{\omega})^{j \color{red}{m}}$<br/>

$\overline{\omega} = \omega^{-1}$
$\implies (\overline{\omega})^{j \color{red}{m}} = ({\omega})^{-j \color{red}{m}}$<br/>
Bringing this term into the summation, we get <br/>
$ \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk}\right)  ({\omega})^{-j \color{red}{m}} =  \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk} ({\omega})^{-j \color{red}{m}} \right)$, because $({\omega})^{-j \color{red}{m}}$ is indepedent of k <br/>
$ =  \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{jk-jm} \right)$<br/>
$ =  \sum_{j=0}^{n-1} \left( \sum_{k=0}^{n-1} a_k \omega^{j(k-m)} \right)$<br/>
Since this is a finite summation, we can change the order of summation <br/>
$ =  \sum_{k=0}^{n-1} \left( \sum_{j=0}^{n-1} a_k \omega^{j(k-m)} \right)$<br/>
This makes $a_k$ a constant inside the second summation, so we can take that out as common <br/>
$ =  \sum_{k=0}^{n-1} a_k \left( \sum_{j=0}^{n-1} \omega^{j(k-m)} \right)$<br/>
Now, lets consider 2 cases:
1. $ k \neq m$, let $ p = k-m $ <br/>
$\sum_{j=0}^{n-1} \omega^{j(k-m)}  = \omega^{0p} + \omega^{1p} + \omega^{2p} + \cdots + \omega^{(n-1)p}$,
which is a geometric series with $\omega^p$ as the common ratio. This summation gives <br/>
$\sum_{j=0}^{n-1} \omega^{j(k-m)} = \frac{1 - \omega^{np}}{1 - \omega^p}$ <br/>
Since, $\omega_n^n = 1 \implies (\omega_n^{np}) = (\omega_n^{n})^{p} = 1$. Thus, $\frac{1 - \omega^{np}}{1 - \omega^p} = 0$<br/>
$ \sum_{k=0}^{n-1} a_k \left( \sum_{j=0}^{n-1} \omega^{j(k-m)} \right) = 0$ <br/>

2. $ k = m$ <br/>
$ \sum_{k=0}^{n-1} a_m \left( \sum_{j=0}^{n-1} \omega^{j(k-m)} \right) = \sum_{k=0}^{n-1} a_m \left( \sum_{j=0}^{n-1} \omega^{0} \right) = \sum_{k=0}^{n-1} a_m  = n a_{\color{red}{m}} $ <br/>
Hence proved.
